```
import pint
import numpy as np
import matplotlib.pyplot as plt

import qtt
import qtt.utilities.visualization
from mpl_toolkits.mplot3d import Axes3D

ureg = pint.UnitRegistry()
```

```
d1 = qtt.data.load_dataset('Repository/Repository/Figure_2/data/22-03-13_qtt_scan2D')
d3 = qtt.data.load_dataset('Repository/Repository/Figure_2/data/06-00-25_qtt_scan2D')
d2 = qtt.data.load_dataset('Repository/Repository/Figure_2/data/02-41-00_qtt_scan2D')
d4 = qtt.data.load_dataset('Repository/Repository/Figure_2/data/09-20-29_qtt_scan2D')
d5 = qtt.data.load_dataset('Repository/Repository/Figure_2/data/11-49-13_qtt_scan2D')
d6 = qtt.data.load_dataset('Repository/Repository/Figure_2/data/18-06-02_qtt_scan2D')
```

```
print(d1,d2,d3,d4,d5,d6)

plt.rcParams.update({'axes.labelsize': 7,
                     'xtick.labelsize': 7,
                     'ytick.labelsize': 7,
                     'legend.fontsize': 16,
                     'font.family': 'sans-serif',
                     'font.sans-serif': ['Tahoma']})


cmap = 'inferno'

x_d1 = ureg.mV*np.array(d1.B1b)
y_d1 = ureg.mV*np.array(d1.B2b)
z_d1 = ureg.pA*(1e12*np.array(d1.Idc2)+12)

x_d2 = ureg.mV*np.array(d2.B1b)
y_d2 = ureg.mV*np.array(d2.B2b)
z_d2 = ureg.pA*(1e12*np.array(d2.Idc2)+12)

x_d3 = ureg.mV*np.array(d3.B1b)
y_d3 = ureg.mV*np.array(d3.B2b)
z_d3 = ureg.pA*(1e12*np.array(d3.Idc2)+12)

x_d4 = ureg.mV*np.array(d4.B1b)
y_d4 = ureg.mV*np.array(d4.B2b)
z_d4 = ureg.pA*(1e12*np.array(d4.Idc2)+12)

x_d5 = ureg.mV*(np.array(d5.B1b)-14)
y_d5 = ureg.mV*(np.array(d5.P1bP2b)*3)
z_d5 = ureg.PA*(1e12*np.array(d5.Idc2)+12)

print(x_d5)
print(y_d5)
print(z_d5)

x_d6 = ureg.mV*np.array(d6.B1b)
y_d6 = ureg.mV*(np.array(d6.P1bP2b)*3)
z_d6 = ureg.pA*(1e12*np.array(d6.Idc2)+12)

print(x_d6)
print(y_d6)
print(z_d6)
```

```

plt.style.use('seaborn-dark')

fszx = 83/25.4 #scale fig 1-4 x size in mm, conv to in 
fszy = 70/25.4 #scale fig 1-4 y size in mm, conv to in 
qdpi = 3400    #png 1-4 dpi

#Graph D1
###########################################
qfiga = plt.figure(num='Figure 2c',
                    figsize=(fszx,
                             fszy),
                    constrained_layout=True)

qplta = plt.pcolormesh(x_d1,y_d1,z_d1,
                      shading='auto',
                      rasterized=True)

plt.xticks([3770,3790,3810,3830,3850,3870],['3770','','','','','3870'])
plt.yticks([3630,3650,3670,3690,3710,3730],['3630','','','','','3730'])

plt.set_cmap(cmap)
plt.ylabel('$G_{5}$ $mV$', labelpad=-13)
plt.xlabel('$G_{3}$ $mV$', labelpad=-6)
plt.title('$G_{4}$ at $1245 mV$')
qfiga.savefig('D1.png', dpi=qdpi)
###########################################

#Graph D2
###########################################
qfiga = plt.figure(num='Figure 2c',
                    figsize=(fszx,
                             fszy),
                    constrained_layout=True)

qplta = plt.pcolormesh(x_d2,y_d2,z_d2,
                      shading='auto',
                      rasterized=True)

plt.xticks([3770,3790,3810,3830,3850,3870],['3770','','','','','3870'])
plt.yticks([3630,3650,3670,3690,3710,3730],['3630','','','','','3730'])

plt.set_cmap(cmap)
plt.ylabel('$G_{5}$ $mV$', labelpad=-13)
plt.xlabel('$G_{3}$ $mV$', labelpad=-6)
plt.title('$G_{4}$ at $1308 mV$')
qfiga.savefig('D2.png', dpi=qdpi)
###########################################

#Graph D3
###########################################
qfiga = plt.figure(num='Figure 2c',
                    figsize=(fszx,
                             fszy),
                    constrained_layout=True)

qplta = plt.pcolormesh(x_d3,y_d3,z_d3,
                      shading='auto',
                      rasterized=True)

plt.xticks([3770,3790,3810,3830,3850,3870],['3770','','','','','3870'])
plt.yticks([3630,3650,3670,3690,3710,3730],['3630','','','','','3730'])


plt.set_cmap(cmap)
plt.ylabel('$G_{5}$ $mV$', labelpad=-13)
plt.xlabel('$G_{3}$ $mV$', labelpad=-6)
plt.title('$G_{4}$ at $1353 mV$')
qfiga.savefig('D3.png', dpi=qdpi)
###########################################

#Graph D4
###########################################
qfiga = plt.figure(num='Figure 2c',
                    figsize=(fszx,
                             fszy),
                    constrained_layout=True)

qplta = plt.pcolormesh(x_d4,y_d4,z_d4,
                      shading='auto',
                      rasterized=True)

plt.xticks([3770,3790,3810,3830,3850,3870],['3770','','','','','3870'])
plt.yticks([3630,3650,3670,3690,3710,3730],['3630','','','','','3730'])

plt.set_cmap(cmap)
plt.ylabel('$G_{5}$ $mV$', labelpad=-13)
plt.xlabel('$G_{3}$ $mV$', labelpad=-6)
plt.title('$G_{4}$ at $1398 mV$')
qfiga.savefig('D4.png', dpi=qdpi)
###########################################
```
